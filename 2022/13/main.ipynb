```python
import ast

def compare(left, right):
    
    if type(left) == int and type(right) == int:
        if right < left:
            return False
        elif right > left:
            return True
    else:
        if type(right) == int:
            right = [right]
        if type(left) == int:
            left = [left]


        i = 0
        while True:
            if i >= len(left) and i >= len(right):
                return None
            if i >= len(left):
                return True
            if i >= len(right):
                return False

            result = compare(left[i], right[i])
            if result == None:
                i += 1
            else:
                return result
        
        
def mergeSort(arr):
    if len(arr) <= 1:
        return arr
    
    mid = len(arr) // 2
    left = mergeSort(arr[:mid])
    right = mergeSort(arr[mid:])
    sortedList = []
    
    while len(left) > 0 and len(right) > 0:
        if compare(left[0], right[0]):
            sortedList.append(left.pop(0))
        else:
            sortedList.append(right.pop(0))
            
    sortedList += left + right
    
    return sortedList

lists = []
with open("data.txt") as f:
    for line in f:
        if line != "\n":
            lists.append(ast.literal_eval(line.strip("\n")))

product = 1
lists = mergeSort(lists)
for i in range(len(lists)):
    if lists[i] in [[[2]],[[6]]]:
        print(i+1)
        product *= (i + 1)
        
print(product)
```
