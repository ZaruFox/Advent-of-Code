```python
import math
 
class File:
    def __init__(self, size, name):
        self.name = name
        self.size = size

    def getSize(self):
        return self.size

class Dir:
    def __init__(self, name, isRoot = False):
        self.name = name
        self.contains = []
        self.isRoot = isRoot

    def getSize(self):
        total = 0
        for item in self.contains:
            total += item.getSize()
        return total

    def part1(self):
        total = 0
        for item in self.contains:
            if isinstance(item, Dir):
                total += item.part1()

        size = self.getSize()
        if size <= 100000:
            total += size
        return total

    def part2(self, lowerBound):
        minimum = math.inf
        for item in self.contains:
            size = item.getSize()
            if isinstance(item, Dir) and size > lowerBound:
                minimum = min((minimum, size, item.part2(lowerBound)))

        return minimum
                

    def addItem(self, item):
        self.contains.append(item)

    def getItem(self, name):
        for i in self.contains:
            if i.name == name:
                return i

root = Dir("/", True)
current = root
callStack = []

with open("data.txt") as f:
    for line in f:
        lineData = line.strip().split(" ")
                
        #commands
        if lineData[0] == "$":
            if lineData[1] == "cd":
                if lineData[2] == "/":
                    continue
                elif lineData[2] == "..":
                    current = callStack.pop()
                else:
                    callStack.append(current)
                    current = current.getItem(lineData[2])

        #list
        else:
            if lineData[0] == "dir":
                current.addItem(Dir(lineData[1]))
            else:
                current.addItem(File(int(lineData[0]), lineData[1]))

print(root.part1())
print(root.part2(30000000 - (70000000 - root.getSize())))
```
