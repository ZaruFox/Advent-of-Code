```python
from copy import deepcopy

elvesPosition = []
map_ = []
directionOrder = ["N", "S", "W", "E"]
with open("data.txt") as f:
    for line in f:
        map_.append(list(line.strip()))
        
for y in range(len(map_)):
    for x in range(len(map_[y])):
        if map_[y][x] == "#":
            elvesPosition.append((x,y))
            
rounds = 0
idleMonkeys = 0

while True:
    rounds += 1
    print(f"Round {rounds}, {idleMonkeys}")
    
    originalElvesPosition = deepcopy(elvesPosition)
    idleMonkeys = 0
    blockedPositions = []
    
    for i, elf in enumerate(elvesPosition):
        checkList = []
        for otherElf in originalElvesPosition:
            if abs(elf[0] - otherElf[0]) <= 1 and abs(elf[1] - otherElf[1]) <= 1:
                checkList.append(otherElf) 
                
        if checkList == [elf]:
                idleMonkeys += 1
        else:
            for direction in directionOrder:
                if direction == "N":
                    if not ((elf[0], elf[1]-1) in checkList or (elf[0]-1, elf[1]-1) in checkList or (elf[0]+1, elf[1]-1) in checkList):
                        elvesPosition[i] = (elvesPosition[i][0], elvesPosition[i][1]-1)
                        break

                elif direction == "S":
                    if not ((elf[0], elf[1]+1) in checkList or (elf[0]-1, elf[1]+1) in checkList or (elf[0]+1, elf[1]+1) in checkList):
                        elvesPosition[i] = (elvesPosition[i][0], elvesPosition[i][1]+1)
                        break

                elif direction == "W":
                    if not ((elf[0]-1, elf[1]) in checkList or (elf[0]-1, elf[1]-1) in checkList or (elf[0]-1, elf[1]+1) in checkList):
                        elvesPosition[i] = (elvesPosition[i][0]-1, elvesPosition[i][1])
                        break

                elif direction == "E":
                    if not ((elf[0]+1, elf[1]) in checkList or (elf[0]+1, elf[1]-1) in checkList or (elf[0]+1, elf[1]+1) in checkList):
                        elvesPosition[i] = (elvesPosition[i][0]+1, elvesPosition[i][1])
                        break
                        
            if len(set(elvesPosition[:i+1])) != len(elvesPosition[:i+1]):
                otherIndex = elvesPosition.index(elvesPosition[i])
                blockedPositions.append(elvesPosition[i])
                elvesPosition[i] = originalElvesPosition[i]
                elvesPosition[otherIndex] = originalElvesPosition[otherIndex]
            elif elvesPosition[i] in blockedPositions:
                elvesPosition[i] = originalElvesPosition[i]
                
    if idleMonkeys == len(elvesPosition):
        print("AAAAAA")
        break
        
    directionOrder.append(directionOrder.pop(0))
    
            
print(rounds)
```
