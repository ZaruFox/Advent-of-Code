```python
from IPython.display import clear_output
import time

wallLocations = []
sandLocations = []
source = [500 , 0]

def printMap(wallLocations,sandLocations, fallingSandLocation = None):
    clear_output(wait=True)
    map_ = ""
    for y in range(100):
        for x in range(424, 550):
            if [x, y] in wallLocations:
                map_  += "#"
            elif fallingSandLocation == [x,y] or [x,y] in sandLocations:
                map_  += "o"
            else:
                map_  += "."
        map_  += "\n"
    print(map_)
    
with open("data.txt") as f:
    for line in f:
        lineData = [[int(x) for x in y.split(",")] for y in line.strip().split(" -> ")]
        
        for i in range(len(lineData) - 1):
            if lineData[i][0] == lineData[i+1][0]:
                for y in range(lineData[i][1], lineData[i+1][1], int((lineData[i][1]-lineData[i+1][1])/-abs(lineData[i][1]-lineData[i+1][1]))):
                    if not [lineData[i][0], y] in wallLocations:
                        wallLocations.append([lineData[i][0], y])
                        
            if lineData[i][1] == lineData[i+1][1]:
                for x in range(lineData[i][0], lineData[i+1][0], int((lineData[i][0]-lineData[i+1][0])/-abs(lineData[i][0]-lineData[i+1][0]))):
                    if not [x, lineData[i][1]] in wallLocations:
                        wallLocations.append([x, lineData[i][1]])
                        
        wallLocations.append([lineData[len(lineData)-1][0], lineData[len(lineData)-1][1]])
      
    
#find max y
biggestY = -1000
for wall in wallLocations:
    if wall[1] > biggestY:
        biggestY = wall[1]
                        
sandCounter = 0
fallingSandLocation = None
while True:
    printMap(wallLocations,sandLocations, fallingSandLocation)
    if fallingSandLocation == None:
        fallingSandLocation = source
        sandCounter += 1
    else:
        if fallingSandLocation[1] > biggestY:
            break
        elif not [fallingSandLocation[0], fallingSandLocation[1]+1] in (wallLocations+sandLocations):
            fallingSandLocation = [fallingSandLocation[0], fallingSandLocation[1]+1]
        elif not [fallingSandLocation[0]-1, fallingSandLocation[1]+1] in (wallLocations+sandLocations):
            fallingSandLocation = [fallingSandLocation[0]-1, fallingSandLocation[1]+1]
        elif not [fallingSandLocation[0]+1, fallingSandLocation[1]+1] in (wallLocations+sandLocations):
            fallingSandLocation = [fallingSandLocation[0]+1, fallingSandLocation[1]+1]
        else:
            sandLocations.append(fallingSandLocation)
            fallingSandLocation = None
            
print(sandCounter-1)
```
